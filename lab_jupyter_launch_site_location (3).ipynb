```python
import folium
from folium.plugins import MarkerCluster

# Create map
spacex_map = folium.Map(
    location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()],
    zoom_start=5
)

# Force clusters to break early
marker_cluster = MarkerCluster(
    disableClusteringAtZoom=6
).add_to(spacex_map)

# Add markers (USE marker_color column)
for _, row in spacex_df.iterrows():
    
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=row['Launch Site'],
        icon=folium.Icon(color=row['marker_color'], icon='info-sign')
    ).add_to(marker_cluster)

spacex_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math
import folium

# Step 1: Distance function
def calculate_distance(lat1, lon1, lat2, lon2):
    return math.sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2)

# Step 2: Launch site location
launch_site_lat = spacex_df.iloc[0]['Lat']
launch_site_lon = spacex_df.iloc[0]['Long']

# Step 3: Target location
target_lat = 28.56367
target_lon = -80.57163

# Step 4: Calculate distance
distance = calculate_distance(launch_site_lat, launch_site_lon, target_lat, target_lon)
print(distance)

# Step 5: Draw line
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [target_lat, target_lon]],
    color='blue'
).add_to(site_map)

# Step 6: Add distance label
folium.Marker(
    location=[target_lat, target_lon],
    icon=folium.DivIcon(
        html=f'<div style="font-size: 12px; color:#d35400;">{distance:.2f}</div>'
    )
).add_to(site_map)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
